```python
## National Treasure Game by Area Galura ##

"""
A) Introduction:
The National Treasure Game is based on one of my favorite movies,"National Treasure". 
You will be playing the game patterned after Nicolas Cage's character as a 
historian, codebreaker and treasure hunter. The game will test if you have
what it takes to steal the Declaration of Independence before our enemies 
or the government stops us. The legendary document will be the first of multiple keys 
that will lead us to the treasures. This game consists of 3 rounds to get 
to the second key, and so on and so forth with "other games". 
There is no skill involved, however, a bit of smart thinking is required.
    
Round 1: Getting access to the location of the Declaration of Independence
Round 2: Fighting off your ex-partner Ian and his associates and/or the guards
Round 3: Decoding the Declaration of Independence to find the second key

B) Known Bugs and/or Errors:
None

"""
```

```python
# Welcome Prompt #
def intro():
    user_name = input (
    """Hello there, Treasure Hunter! Tell me your name before we begin our adventure. """)

    print(f"Welcome {user_name}!")
    input('<Press any key to continue>\n')

    print(f"We have heard a lot about you {user_name}.")
    country = input("Where you are from? \n>")
    fear = input("Your greatest fear? \n>")
    last_adventure = input("Your latest adventure in what place? \n>")

    print(f"""
    Well, {user_name}, you're quite a popular historian, codebreaker and treasure hunter! 
    We heard you are from {country}, your greatest fear is {fear} and you are
    qualified to take on this task based on your adventures in {last_adventure}.\n""")
    
intro()
```

```python
# Round 3 #
def round_3():
    
    print(f"""
We are now in our secret hideout to decode the cypher at the back of The 
Declaration of Independence. The code must be written in invisible ink!
We need to use lemon juice and a hair dryer for the message to appear. 
Your third and final task to decode the location of the second key is to know how
much lemon juice we are going to use. Note that using too much will destroy
The Declaration of Independence, while using too little will have no effect.
    
Now, be careful in your choice, Treasure Hunter. The treasures will forever be
lost if you make a mistake, and we will be arrested if we don't return the
Declaration of Independence in good condition.
    
    1) A slice of lemon 
    2) A whole lemon
    3) A drop of lemon juice
    """)

round_3()   
    
# declaring objects
number = '1' # random number remains the same for each guess
guesses = 3


# looping until player is out of guesses
while guesses > 0:
    choice = input('\n What is your guess (1 to 3)?\n> ')

    # converting choice for numerical comparison
    choice = int(choice)
    
    
    # conditional statement based on user input
    if choice == 1:
        print("\nGood choice! We can now put a few drops to decode the hidden cypher and find the second key!")
        break

    elif choice != number:
        guesses -= 1

        if guesses > 0:
            print(f"\nUh-oh! That's incorrect, you might ruin the Declaration of Independence. You have {guesses} guesses remaining.")

        elif guesses == 0:
            print(f"\nYou have failed, Treasure Hunter!. The number was {number}. Goodbye.")

        else:
            print('Error.')

    # setting a different error message for inner and outer loops
    else:
        print("\nSomething went wrong.")
```
